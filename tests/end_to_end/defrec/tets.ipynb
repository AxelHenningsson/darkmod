```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np



fontsize = 14 # General font size for all text
ticksize= 14 # tick size
plt.rcParams['font.size'] = fontsize
plt.rcParams['xtick.labelsize'] = ticksize
plt.rcParams['ytick.labelsize'] = ticksize
# Data for colormap ranges
ranges = [(-25.58, 25.58), (-2.41, 2.41), (-0.99, 0.99), (-0.4, 0.4)]
labels = ['(d)', '(c)', '(b)', '(a)']
units = r"$\times 10^{-5}$"

# Create the figure
fig, axes = plt.subplots(4, 1, figsize=(12, 3))

# Create colorbars
for i, (vmin, vmax) in enumerate(ranges):
    ax = axes[i]
    ax.axis('off')  # Turn off the axes


    cmap = plt.cm.Spectral
    norm = mcolors.TwoSlopeNorm(vmin=vmin, vcenter=0, vmax=vmax)
    
    # Create a colorbar
    cbar = fig.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap),
                        ax=ax,
                        orientation='horizontal', fraction=1, pad=0.2, aspect=40)
    
    # Set label above each colorbar
    cbar.ax.text(-0.1, 0.5, labels[i], ha='center', va='center', fontsize=14, transform=cbar.ax.transAxes)

    #cbar.ax.set_xticks([])
    cbar.ax.set_yticks([])

    cbar.ax.spines['top'].set_visible(False)
    cbar.ax.spines['bottom'].set_visible(False)
    cbar.ax.spines['left'].set_visible(False)
    cbar.ax.spines['right'].set_visible(False)
    

    # Add range text at the ends of each colorbar
    cbar.ax.text(-0.04, 0.5, f"{vmin:.2f}", ha='center', va='center', fontsize=14, transform=cbar.ax.transAxes)
    cbar.ax.text(1.03, 0.5, f"{vmax:.2f}", ha='center', va='center', fontsize=14, transform=cbar.ax.transAxes)


# Add overall title for the colorbars
fig.suptitle(f"            Colormap Ranges for zones (a)-(d) in units of {units}", fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```
